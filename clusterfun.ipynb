```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
codes = ['sq', 'mk', 'bg', 'bs', 'hr', 'sr', 'el', 'ro']
langs = ['albanian', 'macedonian', 'bulgarian', 'bosnian', 
         'croatian', 'serbian', 'greek', 'romanian']
```

```python
f = open('data/chunks.pkl', 'rb')
df = pickle.load(f)
f.close()
labels = sorted(df['lang'].unique().tolist())
df = None
```

```python
labels[:5]
```

```python
indices = [labels.index(l) for l in codes]
indices
```

```python
f = open('matrix_nonanonNBbis.pkl', 'rb')
allmat = pickle.load(f)
f.close()
```

```python
mat = allmat[indices][:, indices]
mat
```

```python
f = open('matrix_anonNBbis25.pkl', 'rb')
allanon = pickle.load(f)
f.close()
```

```python
anon = allanon[indices][:, indices]
anon
```

```python
fig, ax = plt.subplots(figsize=(10,10))         
sns.heatmap(mat, square=True, annot=True , cmap="Reds",
           xticklabels=langs, yticklabels=langs, ax=ax)
plt.savefig('balkansNonanon.png')
plt.show()
```

These languages are more commonly confused with each other than the Balkan Sprachbund ones, likely in no small part because they all use the Latin writing system. Romanian was confused the most as being Latin (with Italian close behind), while Latin was by far confused the most as being Italian. Catalan was confused almost four times more with Spanish than with French. French was mistaken the most often for Italian. 

```python
fig, ax = plt.subplots(figsize=(10,10))         
sns.heatmap(anon, square=True, annot=True , cmap="Reds",
           xticklabels=langs, yticklabels=langs, ax=ax)
plt.savefig('balkansAnon.png')
plt.show()
```
